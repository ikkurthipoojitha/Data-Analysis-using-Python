### Identifying and eliminating outliers:
- outliers are observations that are significantly different from other data points
- outliers can affect the training process of a machine learning algorithm, resulting in a loss of accuracy
- IQR = Q3(quantile(0.75)) - Q1(quantile(0.25))

## Identifying and Eliminating outliers

```python
adv = pd.read_csv("https://raw.githubusercontent.com/ikkurthipoojitha/Data-Analysis-using-Python/main/Datasets/Advertising.csv")
adv.head()
```

```python
import matplotlib.pyplot as plt
```

```python
adv.plot(kind = "box")
plt.show()
```

```python
Q3 = adv.quantile(0.75)
Q1 = adv.quantile(0.25)
IQR = Q3 - Q1
IQR
```

```python
filter_data = adv [((adv<(Q1 - 1.5 * IQR))|(adv>(Q3 + 1.5 * IQR))).any(axis=1)]
filter_data
```

```python
Q3 = adv.quantile(0.75)
Q1 = adv.quantile(0.25)
IQR = Q3 - Q1
IQR
filter_data = adv[~((adv<(Q1 - 1.5 * IQR))|(adv>(Q3 + 1.5 * IQR))).any(axis = 1)]
filter_data
```

```python
filter_data.plot(kind = "box")
plt.show()
```
